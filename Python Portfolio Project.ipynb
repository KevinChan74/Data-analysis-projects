```python
# Read in the data

df = pd.read_csv('/Users/hickeychan/Desktop/movies.csv')
```

```python
# Copy the data; ds = dataset

ds = df.copy()
```

```python
# Look at the data

ds
```

```python
# Check if there is any missing data

for col in ds.columns:
    pct_missing = np.mean(ds[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Remove duplicates: Only rows with same data in every columns would be removed

ds.drop_duplicates()
```

```python
# Are there any outliers?

ds.boxplot(column = ['gross'])
```

```python
# Order data

ds.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Show rows with scrollable interface in Jupyter Notebook: pd.set_option('display.max_rows', None)
```

```python
# Hypothesis: Year has high correlation with Gross Earnings; Company high correlation with Gross Earnings
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = ds['budget'], y = ds['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
ds.apply(lambda x: x.factorize()[0]).corr(method = 'pearson')
```

```python
# Change categorical data into numeric data to get correlation matrix


corrlation_matrix = ds.apply(lambda x: x.factorize()[0]).corr(method = 'pearson')
sns.heatmap(corrlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
correlation_matrix = ds.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_matrix.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# See which ones have high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
